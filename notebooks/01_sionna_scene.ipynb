```python
import numpy as np
from sionna.rt import load_scene, PlanarArray, Transmitter, RadioMapSolver
scene = load_scene("munich")
scene.tx_array = PlanarArray(num_rows=1,num_cols=1,pattern="iso",polarization="V")
scene.rx_array = scene.tx_array
scene.add(Transmitter("tx", position=[8.5,21,27]))
rm = RadioMapSolver()(scene, max_depth=5, cell_size=(1.0,1.0), samples_per_tx=10**6)
gain = 10*np.log10(np.maximum(rm.path_gain.numpy()[0], 1e-30))
np.save("sionna_gain_grid.npy", gain)
print("grid", gain.shape, "dB range", float(gain.min()), float(gain.max()))
from google.colab import files; files.download("sionna_gain_grid.npy")
```
